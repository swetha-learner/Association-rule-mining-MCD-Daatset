## MC Donald dataset 

```python
import pandas as pd
df = pd.read_excel('.\data\mcd.xlsx',header = None)
df.head()
```

```python
df.drop(columns=[0], inplace = True)
df
```

```python
dataset = [[element for element in list1 if str(element)!= 'nan' ] for list1 in df.values]
dataset[:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te=te.fit(dataset) #reads all unique values in your list
print(te.columns_)
te_ary = te.fit_transform(dataset) #creates a boolean 2d array from dataset
print(te_ary)
```

```python
mcddf = pd.DataFrame(te_ary, columns = te.columns_)
mcddf.head()
```

```python
from mlxtend.frequent_patterns import apriori

#finds frequency of items or itemsets
frequent_itemsets = apriori(mcddf, use_colnames = True, min_support = 0.15)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns  import association_rules

#Forming asso rrules amd returning all performances metrics of those asso
rules = association_rules(frequent_itemsets,metric ='support',min_threshold=0.15)
rules
```

```python
newrules = rules[rules['confidence']>0.15]
newrules
```
